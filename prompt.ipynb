**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? **Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
data["coupon"].unique()
bar_coupon_data = data[data["coupon"] == "Bar"]
print(type(bar_coupon_data))
print(bar_coupon_data.shape)
bar_coupon_data.head()
```

2. What proportion of bar coupons were accepted?


```python

bar_coupon_acceptance_rate = bar_coupon_data['Y'].mean()
print("Proportion of bar coupons accepted:", bar_coupon_acceptance_rate)

#sanity check
print(bar_coupon_data['Y'].value_counts(normalize=True))
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
low_freq_bar_visitors = ["never", "less1", "1~3"]
high_freq_bar_visitors = ["4~8", "gt8"]
low_freq_bar_dataframe = bar_coupon_data[
    bar_coupon_data["Bar"].isin(low_freq_bar_visitors)
]
high_freq_bar_dataframe = bar_coupon_data[
    bar_coupon_data["Bar"].isin(high_freq_bar_visitors)
]
low_freq_bar_acceptance_rate = low_freq_bar_dataframe['Y'].mean()
high_freq_bar_acceptance_rate = high_freq_bar_dataframe['Y'].mean()
print("Low frequency bar visitors acceptance rate:", low_freq_bar_acceptance_rate)
print("High frequency bar visitors acceptance rate:", high_freq_bar_acceptance_rate)
summary = pd.DataFrame({
    "bar_visit_group": ["3 or fewer per month", "More than 3 per month"],
    "acceptance_rate": [low_freq_bar_acceptance_rate, high_freq_bar_acceptance_rate],
    "sample_size": [low_freq_bar_dataframe.shape[0], high_freq_bar_dataframe.shape[0]]
})
summary.head()
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
more_than_once_a_month = ["1~3", "4~8", "gt8"]
bar_coupon_data_norm = bar_coupon_data.copy()
bar_coupon_data_norm['age_num'] = pd.to_numeric(bar_coupon_data['age'], errors='coerce')
target_group = bar_coupon_data_norm[
    (bar_coupon_data_norm['Bar'].isin(more_than_once_a_month)) &
    (bar_coupon_data_norm['age_num'] >= 25)
]
comparison_group = bar_coupon_data_norm.drop(target_group.index)
target_group_acceptance_rate = target_group['Y'].mean()
comparison_group_acceptance_rate = comparison_group['Y'].mean()
print("Target group acceptance rate:", target_group_acceptance_rate)
print("Comparison group acceptance rate:", comparison_group_acceptance_rate)
summary_2 = pd.DataFrame({
    "group": ["Target Group (Frequent Bar Visitors Age 25+)", "Comparison Group"],
    "acceptance_rate": [target_group_acceptance_rate, comparison_group_acceptance_rate],
    "sample_size": [target_group.shape[0], comparison_group.shape[0]]
})
summary_2.head()
print(
    'There is a difference between this result and the previous one \n' \
    'because the target group in this analysis includes only individuals \n' \
    'aged 25 and older who visit bars more than once a month, \n' \
    'while the previous analysis grouped all frequent bar visitors together regardless of age. \n' \
    'By adding the age criterion, we may be capturing a different demographic that has distinct \n' \
    'coupon acceptance behaviors.'
)
```
